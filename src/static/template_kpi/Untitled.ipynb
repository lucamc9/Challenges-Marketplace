### Yearly data

```python
year_data = excel.parse(excel.sheet_names[0])
```

```python
# Sales concentration
sales_names = []
sales_15 = []
sales_16 = []
sales_17 = []
for i in range(3, 9):
    sales_names.append(year_data['Unnamed: 2'].iloc[i])
    sales_15.append(year_data['FY 2015'].iloc[i])
    sales_16.append(year_data['FY 2016'].iloc[i])
    sales_17.append(year_data['FY 2017'].iloc[i])
    if year_data['Unnamed: 2'].iloc[i] == 'All Other Customers':
        break
```

```python
# Sales, net income, cash
yearly_sales = []
total_sales = year_data[year_data['Unnamed: 2'] == 'Total Sales']
yearly_sales.append(total_sales.iloc[0]['FY 2015'])
yearly_sales.append(total_sales.iloc[0]['FY 2016'])
yearly_sales.append(total_sales.iloc[0]['FY 2017'])
```

```python
# Gross profit margin
gross = []
gross_margin = year_data[year_data['Unnamed: 2'] == 'Profit Margin %']
gross.append(gross_margin.iloc[0]['FY 2015'])
gross.append(gross_margin.iloc[0]['FY 2016'])
gross.append(gross_margin.iloc[0]['FY 2017'])
# Net profit margins
net = []
net_margin = year_data[year_data['Unnamed: 2'] == 'Net Profit Margin %']
net.append(net_margin.iloc[0]['FY 2015'])
net.append(net_margin.iloc[0]['FY 2016'])
net.append(net_margin.iloc[0]['FY 2017'])
```

```python
# Cash in Bank
cash_bank = []
cashes_bank = year_data[year_data['Unnamed: 1'] == 'Cash & Equivalents']
cash_bank.append(cashes_bank.iloc[0]['FY 2015'])
cash_bank.append(cashes_bank.iloc[0]['FY 2016'])
cash_bank.append(cashes_bank.iloc[0]['FY 2017'])
# Accounts Receivable
accounts_receivable = []
receivables = year_data[year_data['Unnamed: 1'] == 'Accounts Receivable']
accounts_receivable.append(receivables.iloc[0]['FY 2015'])
accounts_receivable.append(receivables.iloc[0]['FY 2016'])
accounts_receivable.append(receivables.iloc[0]['FY 2017'])
# Accounts Payable
accounts_payable = []
payable = year_data[year_data['Unnamed: 1'] == 'Accounts Payable']
accounts_payable.append(payable.iloc[0]['FY 2015'])
accounts_payable.append(payable.iloc[0]['FY 2016'])
accounts_payable.append(payable.iloc[0]['FY 2017'])
```

```python
# Quick Ratio
quick_ratio = []
current_liabilities = []
liabilities = year_data[year_data['Unnamed: 1'] == 'Current Liabilities']
current_liabilities.append(liabilities.iloc[0]['FY 2015'])
current_liabilities.append(liabilities.iloc[0]['FY 2016'])
current_liabilities.append(liabilities.iloc[0]['FY 2017'])
for i in range(len(current_liabilities)):
    quick_ratio.append((cash_bank[i] + accounts_receivable[i])/current_liabilities[i])
    
# Current Ratio
current_ratio = []
current_assets = []
assets = year_data[year_data['Unnamed: 1'] == 'TOTAL CURRENT ASSETS']
current_assets.append(assets.iloc[0]['FY 2015'])
current_assets.append(assets.iloc[0]['FY 2016'])
current_assets.append(assets.iloc[0]['FY 2017'])
for i in range(len(current_assets)):
    current_ratio.append(current_assets[i]/current_liabilities[i])
```

```python
# Inventory turns
cost_goods_yearly = []
inventory_yearly = []
goods = year_data[year_data['Unnamed: 1'] == 'Cost of Goods Sold']
cost_goods_yearly.append(goods.iloc[0]['FY 2015'])
cost_goods_yearly.append(goods.iloc[0]['FY 2016'])
cost_goods_yearly.append(goods.iloc[0]['FY 2017'])
inventory = year_data[year_data['Unnamed: 1'] == 'Inventory']
inventory_yearly.append(inventory.iloc[0]['FY 2015'])
inventory_yearly.append(inventory.iloc[0]['FY 2016'])
inventory_yearly.append(inventory.iloc[0]['FY 2017'])
inventory_turns = []
for i in range(len(cost_goods_yearly)):
    inventory_turns.append(cost_goods_yearly[i]/inventory_yearly[i])
    
```
